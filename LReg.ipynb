```python
 from sklearn.linear_model import LinearRegression
```

```python
x = np.array([5, 15, 25, 35, 45, 55 ]).reshape(-1,1)
```

```python
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
r_sq = model.score(x, y)
```

```python
print(f"final result {r_sq}")
```

```python
 print(f"slope: {model.coef_}")
    
```

```python
 print(f"intercept: {model.intercept_}")
```

```python
plt.plot(x)
```

```python
plt.plot(y)
```

```python
plt.plot(x,y,'*')
```

```python
var =[]
for i in range (len(x)):
    var.append(  model.coef_* x[i] + model.intercept_)
```

```python
plt.plot(var)
```

```python
plt.plot(x,y,'*')
plt.plot(x,var)
```

```python
 y_pred = model.predict(x)
```

```python
print(f"model prrdiction{y_pred}")
```

```python
plt.plot(y_pred)
plt.plot(y)
```

```python
plt.plot(x,y_pred)
plt.plot(x,var)
```
