```python
# we come to know that income is depenedent on two independent variables experience and age .
X = df.drop('income',axis = 1)
```

```python
print(X)
```

```python
Y = df['income']
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test, = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
print(X_train)
print("\n")
print(Y_train)
print("\n")
print(X_test)
print("\n")
print(Y_test)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.plot(X_test,Y_pred)
plt.plot(X_test,Y_pred,"*")
plt.plot(X_test,Y_test)
plt.plot(X_test,Y_test,"^")
```
